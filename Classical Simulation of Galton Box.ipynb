```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
#Parameters
n_levels = 10  #Numbers of levels (pegs)
n_particles = 1000 #Number of balls dropped

#Simulation: each ball randomly moves to the left or right at each level
positions = []
for i in range(n_particles):
    right_moves = 0
    for i in range(n_levels):
        if np.random.rand() > 0.5: # Right
            right_moves += 1
    positions.append(right_moves)

# Count final bin positions
counts = Counter(positions)

#Plot Histogram
bins = range(n_levels + 1)
frequencies = [counts.get(bin, 0) for bin in bins]

plt.figure(figsize=(10,6))
plt.bar(bins, frequencies, color="skyblue", edgecolor="black")
plt.title(f'Classical Galton Simulator(n={n_levels}, n_particles={n_particles})')
plt.xlabel("Final Bin (Number of right moves)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()






```
